Напишите программу реализующую Алгоритм Евклида


> Алгоритм Евклида – это алгоритм нахождения наибольшего общего делителя (НОД) пары целых чисел.



Ввод:
```
30 18
```

Вывод:

```
6
```

Этап 2:
Оформите ваш код в виде функции


````python
def n(s1,s2):
  if s1>s2:
    n=1
    for x in range(1,s2+1):
      if s1%x==0 and s1%x==0 and n<x:
        n=x
    return n
  elif s2>s1:
    n=1
    for x in range(1,s1+1):
      if s2%x==0 and s1%x==0:
        n=x
    return n
  else:
    return 1


print('Введите два числа:')
s1=int(input())
s2=int(input())
n=n(s1,s2)
print(n)
````

**Задача 4:**

Этап 1:
Напишите функцию программу, которая принимает строку и возвращает список слов и количество их упомнинаний в предложении

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
apple banana apple
```

Вывод:

```
apple: 2,
banana: 1
```

````python
def g(pr):
    w=[]
    j=[]
    pr = pr.split(' ')
    for x in pr:
        if x not in w:
            w.append(x)
    for i in w:
        l = 0
        for k in pr:
            if k == i:
                l += 1
        j.append(i)
        j.append(':')
        j.append(str(l))
    return j


print('Введите предложение:')
pr=input()
print(' '.join(g(pr)))
````

Шифр ​​Цезаря

Напишите программу на Python, которая реализует шифр Цезаря, простой метод шифрования, который заменяет каждую букву буквой на фиксированное количество позиций вниз по алфавиту. Программа должна запрашивать у пользователя сообщение и значение сдвига, а затем шифровать и расшифровывать сообщение.

Этап 1:

Напишите код для реализации данной задачи

Этап 2:

Оформите код в виде нескольких функций:

* Зашифровывает сообщение
* Расшифровывает сообщение

````python
def ras(soo,sd):
    k=['а', 'б', 'в', 'г', 'д', 'е', 'ё', 'ж', 'з', 'и', 'й', 'к', 'л', 'м', 'н', 'о', 'п', 'р', 'с', 'т', 'у', 'ф', 'х', 'ц', 'ч', 'ш', 'щ', 'ъ', 'ы', 'ь', 'э', 'ю', 'я']
    newsoo=[]
    for x in soo:
        poz=-1
        for y in k:
            poz+=1
            w=poz-sd
            if x==y:
                if w<0:
                    w=w+33
                    newsoo.append(k[w])
                else:
                    newsoo.append(k[w])
    return ''.join(newsoo)

def shif(soo,sd):
    k=['а', 'б', 'в', 'г', 'д', 'е', 'ё', 'ж', 'з', 'и', 'й', 'к', 'л', 'м', 'н', 'о', 'п', 'р', 'с', 'т', 'у', 'ф', 'х', 'ц', 'ч', 'ш', 'щ', 'ъ', 'ы', 'ь', 'э', 'ю', 'я']
    newsoo=[]
    for x in soo:
        poz=-1
        for y in k:
            poz+=1
            w=poz+sd
            if x==y:
                if w>32:
                    w-=33
                    newsoo.append(k[w])
                else:
                    newsoo.append(k[w])
    return ''.join(newsoo)


print('Введите сообщение БЕЗ ЗНАКОВ ПРЕПИНАНИЯ:')
sfo=input()
print('Введите размер сдвига:')
sd=int(input())
print('Введите номер команды(1.Расшифровать 2. Зашифровать)')
com=int(input())
sfo1=sfo.split()
sfoig=[]
if com==1:
    for x in sfo1:
        sfoig.append(ras(x,sd))
else:
    for x in sfo1:
        sfoig.append(shif(x,sd))
print(' '.join(sfoig))
````
